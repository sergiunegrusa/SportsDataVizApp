```python
def getTeamPasses(teamId):
    df = pd.read_csv("europaFinal.csv")
    df = df.loc[df['teamId'] == teamId]
    df = df.loc[df['type'] == 'Pass']
    df = df.loc[df['outcome'] == 'Successful']
    return df
```

```python
def getClearance(teamId):
    df = pd.read_csv("europaFinal.csv")
    df = df.loc[df['teamId'] == teamId]
    df = df.loc[df['type'] == 'Clearance']
    return df
```

```python
def teamPasses(teamId):
    df = getTeamPasses(teamId)
    df2 = getClearance(teamId)
    frames = [df, df2]
    result = pd.concat(frames)
    return result
```

```python
def plotPasses(teamId):
    passes = teamPasses(teamId)
    passes.x = passes.x * 1.2
    passes.y = passes.y * 0.8
    passes.endX = passes.endX * 1.2
    passes.endY = passes.endY * 1.2

    passes['beginning'] = np.sqrt(np.square(120-passes['x']) + np.square(80-passes['y']))
    passes['end'] = np.sqrt(np.square(120-passes['endX']) + np.square(80-passes['endY']))

    pitch = Pitch(pitch_type='statsbomb')
    fig,ax = pitch.draw()
    pitch.lines(passes.x, passes.y, passes.endX, passes.endY, comet=True, ax=ax)

plotPasses('Villarreal')
```

```python
def getSidePercentage(x_data, y_data, endX_data, endY_data, value1, value2):
    sum1 = 0
    sum2 = 0
    sum3 = 0
    results = []
    for y in y_data:
        if (0 <= y) & (y < value1):
            sum1 += 1
        if (value1 <= y) & (y < value2):
            sum2 += 1
        if (value2 <= y) & (y < 80):
            sum3 += 1

    sum = sum1 + sum2 + sum3
    results.append(sum1/sum)
    results.append(sum2/sum)
    results.append(sum3/sum)

    return results
```

```python
def pairZones(teamId):
    s = 20
    pitch = Pitch()
    fig,ax = pitch.draw()
    df = pd.read_csv("europaFinal.csv")
    df = df.loc[df['teamId'] == teamId]
    x = df.x*1.2
    y = df.y*0.8
    endX = df.loc[df['outcome'] == 'Successful']['endX']*1.2
    endY = df.loc[df['outcome'] == 'Successful']['endY']*0.8

    zones = getSidePercentage(x, y, endX, endY, 25, 55)
    # print(zones)
    zone1 = zones[0]
    zone2 = zones[1]
    zone3 = zones[2]

    pitch.lines(0, 25 , 119, 25, comet=False, transparent=True, ax=ax)
    pitch.lines(0, 55 , 119, 55, comet=False, transparent=True, ax=ax)

    for zone in zones:
        # print(str(zone) + ' ' + str(type(zone)))
        fillPercentage(zone, s)
        s = s + 20



pairZones('Manchester United')
```

```python
def playerInterestZone(teamId, playerId):
    df = pd.read_csv("europaFinal.csv")
    df = df.loc[df['teamId'] == teamId]
    player = df.loc[(df['playerId'] == playerId)]
    defpoints = player[['x', 'y']].values
    print(defpoints)

    #Create a convex hull object and assign it to the variable hull
    hull = ConvexHull(player[['x','y']])
    print(hull)

    #Plot the X & Y location with dots
    plt.plot(player.x,player.y, 'o')

    #Loop through each of the hull's simplices
    for simplex in hull.simplices:
        #Draw a black line between each
        plt.plot(defpoints[simplex, 0], defpoints[simplex, 1], 'k-')

    #Fill the area within the lines that we have drawn
    plt.fill(defpoints[hull.vertices,0], defpoints[hull.vertices,1], 'k', alpha=0.3)

    # pitch = Pitch()
    # fig, ax = pitch.draw()
    # hull = pitch.convexhull(x, y)
    # poly = pitch.polygon(hull, ax=ax, edgecolor='cornflowerblue', facecolor='cornflowerblue', alpha=0.3)
    # scatter = pitch.scatter(x, y, ax=ax, edgecolor='black', facecolor='cornflowerblue')
    # plt.show()

playerInterestZone('Villarreal', 9)
```

```python
def teamInterestZone(teamId):
    df = pd.read_csv("europaFinal.csv")
    team = df.loc[df['teamId'] == teamId]
    players = team.loc[team["playerId"] != None]
    players = players.playerId.unique()
    print(players)

    #For each player in our players variable
    for player in players:
        #Create a new dataframe for the player
        df = team[(team.playerId == player)]

        #Create an array of the x/y coordinate groups
        points = df[['x', 'y']].values

        #If there are enough points for a hull, create it. If there's an error, forget about it
        try:
            hull = ConvexHull(df[['x','y']])

        except:
            pass

        #If we created the hull, draw the lines and fill with 5% transparent red. If there's an error, forget about it
        try:
            for simplex in hull.simplices:
                plt.plot(points[simplex, 0], points[simplex, 1], 'k-')
                plt.fill(points[hull.vertices,0], points[hull.vertices,1], 'red', alpha=0.05)

        except:
            pass

    #Once all of the individual hulls have been created, plot them together
    plt.show()

teamInterestZone('Villarreal')
```

```python
def makeData(passes, tackles, fouls, unsuccesPass, unsuccesShots, wonAerials, goals):
    data = {'Passes': passes, 'Tackles': tackles, 'Fouls': fouls, 'Unsuccesful Passes':unsuccesPass, "Unsuccesful Shots": unsuccesShots, "Won Aerials": wonAerials, "Goals": goals}
    return data
```

```python
def countElement(array, element):
    sum = 0
    for value in array:
        if value == element:
            sum = sum + 1

    return sum
```

```python
def getBoxCoordinates(teamId, x_coordinate, y_coordinate):
    passes = getTeamPasses(teamId)
    passes.x = passes.x * 1.2
    passes.y = passes.y * 0.8
    passes.endX = passes.endX * 1.2
    passes.endY = passes.endY * 1.2
    passes = passes[passes['x'] > x_coordinate]
    passes = passes[passes['y'] > y_coordinate]
    passes = passes[passes['y'] < (80 - y_coordinate)]
    return passes

x = getBoxCoordinates('Villarreal', 101, 17.0)
print(x)
```

```python
def getAngleWithSin(x, y, endX, endY):
    z = x
    endZ = endY
    ipo = (endY - y)*(endY - y) + (endX - x)*(endX - x)
    ipo = np.sqrt(ipo)
    cat1 = (endZ - y)*(endZ - y) + (z - x)*(z - x)
    cat1 = np.sqrt(cat1)
    sin = cat1/ipo
    angle = math.asin(sin)*180/math.pi
    return angle
```

```python
def getAngleWithCos(x, y , endX, endY, xz, yz):
    cat = (yz - y)*(yz - y) + (xz - x)*(xz - x)
    cat = np.sqrt(cat)

    ipo = (endY - y)*(endY - y) + (endX - x)*(endX - x)
    ipo = np.sqrt(ipo)

    cos = cat/ipo
    angle = math.acos(cos)*180/math.pi
    return angle
```
